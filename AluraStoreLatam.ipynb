### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#

# Calculate total price per store
total_price_tienda1 = tienda['Precio'].sum()
total_price_tienda2 = tienda2['Precio'].sum()
total_price_tienda3 = tienda3['Precio'].sum()
total_price_tienda4 = tienda4['Precio'].sum()

print(f"Total price for tienda 1: {total_price_tienda1}")
print(f"Total price for tienda 2: {total_price_tienda2}")
print(f"Total price for tienda 3: {total_price_tienda3}")
print(f"Total price for tienda 4: {total_price_tienda4}")
```

# 2. Ventas por categoría

```python
#ventas por Categoría del Producto por tienda

ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index()
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index()

print("Ventas por Categoría del Producto en Tienda 1:")
print(ventas_categoria_tienda1)
print("\nVentas por Categoría del Producto en Tienda 2:")
print(ventas_categoria_tienda2)
print("\nVentas por Categoría del Producto en Tienda 3:")
print(ventas_categoria_tienda3)
print("\nVentas por Categoría del Producto en Tienda 4:")
print(ventas_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
#Calificación promedio por tienda

prom_calificacion_tienda = tienda['Calificación'].mean()
prom_calificacion_tienda2 = tienda2['Calificación'].mean()
prom_calificacion_tienda3 = tienda3['Calificación'].mean()
prom_calificacion_tienda4 = tienda4['Calificación'].mean()

print(f"Promedio Calificación Tienda 1: {prom_calificacion_tienda}")
print(f"Promedio Calificación Tienda 2: {prom_calificacion_tienda2}")
print(f"Promedio Calificación Tienda 3: {prom_calificacion_tienda3}")
print(f"Promedio Calificación Tienda 4: {prom_calificacion_tienda4}")
```

# 4. Productos más y menos vendidos


```python
# Productos más vendidos por tienda
productos_mas_vendidos_tienda1 = tienda.groupby('Producto')['Precio'].sum().nlargest(3)
productos_mas_vendidos_tienda2 = tienda2.groupby('Producto')['Precio'].sum().nlargest(3)
productos_mas_vendidos_tienda3 = tienda3.groupby('Producto')['Precio'].sum().nlargest(3)
productos_mas_vendidos_tienda4 = tienda4.groupby('Producto')['Precio'].sum().nlargest(3)

print("Productos más vendidos en Tienda 1:")
print(productos_mas_vendidos_tienda1)
print("\nProductos más vendidos en Tienda 2:")
print(productos_mas_vendidos_tienda2)
print("\nProductos más vendidos en Tienda 3:")
print(productos_mas_vendidos_tienda3)
print("\nProductos más vendidos en Tienda 4:")
print(productos_mas_vendidos_tienda4)
```

```python
# Productos menos vendidos por tienda
productos_menos_vendidos_tienda1 = tienda.groupby('Producto')['Precio'].sum().nsmallest(3)
productos_menos_vendidos_tienda2 = tienda2.groupby('Producto')['Precio'].sum().nsmallest(3)
productos_menos_vendidos_tienda3 = tienda3.groupby('Producto')['Precio'].sum().nsmallest(3)
productos_menos_vendidos_tienda4 = tienda4.groupby('Producto')['Precio'].sum().nsmallest(3)

print("\nProductos menos vendidos en Tienda 1:")
print(productos_menos_vendidos_tienda1)
print("\nProductos menos vendidos en Tienda 2:")
print(productos_menos_vendidos_tienda2)
print("\nProductos menos vendidos en Tienda 3:")
print(productos_menos_vendidos_tienda3)
print("\nProductos menos vendidos en Tienda 4:")
print(productos_menos_vendidos_tienda4)
```

# 5. Envío promedio por tienda

```python
#Costo de envío promedio por tienda

prom_envio_tienda = tienda['Costo de envío'].mean()
prom_envio_tienda2 = tienda2['Costo de envío'].mean()
prom_envio_tienda3 = tienda3['Costo de envío'].mean()
prom_envio_tienda4 = tienda4['Costo de envío'].mean()

print("Costo de envío promedio por tienda")
print(f"Prom Tienda 1: {prom_envio_tienda}")
print(f"Prom Tienda 2: {prom_envio_tienda2}")
print(f"Prom Tienda 3: {prom_envio_tienda3}")
print(f"Prom Tienda 4: {prom_envio_tienda4}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Gráfico de facturación total por tienda
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
total_prices = [total_price_tienda1, total_price_tienda2, total_price_tienda3, total_price_tienda4]

plt.figure(figsize=(8, 5))
sns.barplot(x=stores, y=total_prices)
plt.title('Facturación Total por Tienda')
plt.ylabel('Facturación Total')
plt.xlabel('Tienda')
plt.show()
```

```python
# 2. Gráficos de ventas por categoría para cada tienda (usando bucle para simplificar)

tiendas_df = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, (df, nombre) in enumerate(zip(tiendas_df, nombres_tiendas)):
    ventas_categoria = df.groupby('Categoría del Producto')['Precio'].sum().reset_index()
    sns.barplot(x='Categoría del Producto', y='Precio', data=ventas_categoria, ax=axes[i])
    axes[i].set_title(f'Ventas por Categoría en {nombre}')
    axes[i].set_ylabel('Ventas Totales')
    axes[i].set_xlabel('Categoría del Producto')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 3. Gráfico de calificación promedio por tienda
plt.figure(figsize=(8, 5))
sns.barplot(x=nombres_tiendas, y=[prom_calificacion_tienda, prom_calificacion_tienda2, prom_calificacion_tienda3, prom_calificacion_tienda4])
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Calificación Promedio')
plt.xlabel('Tienda')
plt.show()
```

```python
# 3. Gráfico de costo de envío promedio por tienda
avg_shipping = [prom_envio_tienda, prom_envio_tienda2, prom_envio_tienda3, prom_envio_tienda4]

plt.figure(figsize=(8, 5))
sns.barplot(x=stores, y=avg_shipping)
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.xlabel('Tienda')
plt.show()
```
